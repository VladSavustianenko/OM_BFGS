## Код програми

```python
import numpy as np
import matplotlib.pyplot as plt

global counter
counter = 0
x0 = [-1.2, 0]

h = 0.000001
step = 0.1

e_golden_ratio = 1e-5
e_bfsg = 1e-4
R = 0
```

```python
def func(p):
    global counter
    counter += 1
    return (10 * (p[0] ** 2 - p[1]) ** 2 + (p[0] - 1) ** 2) ** 0.25


def func_alpa(p, s, alpha):
    return func([p[0] + s[0] * alpha, p[1] + s[1] * alpha])


def d_center(p):
    x = (func([p[0] + h, p[1]]) - func([p[0] - h, p[1]])) / (2 * h)
    y = (func([p[0], p[1] + h]) - func([p[0], p[1] - h])) / (2 * h)
    return np.array([x, y])


def d_left(p):
    f = func(p)
    x = (f - func([p[0] - h, p[1]])) / h
    y = (f - func([p[0], p[1] - h])) / h
    return np.array([x, y])


def d_right(p):
    f = func(p)
    x = (func([p[0] + h, p[1]]) - f) / h
    y = (func([p[0], p[1] + h]) - f) / h
    return np.array([x, y])


def norm_x(p):
    return (p[0] ** 2 + p[1] ** 2) ** 0.5
```

```python
def svenn(p, s):
    alpha_array = []
    f_array = []
    direction = 0
    step_svenn = step * norm_x(p) / norm_x(s)

    f1 = func_alpa(p, s, -step_svenn)
    f2 = func_alpa(p, s, 0)
    f3 = func_alpa(p, s, step_svenn)

    alpha_array.append(0)
    f_array.append(f2)
    if f3 > f2 > f1:
        direction = -1
        alpha_array.append(-step_svenn)
        f_array.append(f1)
    elif f3 < f2 < f1:
        direction = 1
        alpha_array.append(step_svenn)
        f_array.append(f3)
    else:
        return [-step_svenn, step_svenn]

    i = 1
    while f_array[i] < f_array[i-1]:
        alpha = alpha_array[i] + 2**i * step_svenn * direction
        alpha_array.append(alpha)
        f_i = func_alpa(p, s, alpha)
        f_array.append(f_i)
        i += 1

    list4_alpha = [alpha_array[i-2], alpha_array[i-1], (alpha_array[i-1] + alpha_array[i]) / 2 ,alpha_array[i]]
    list4_f = [f_array[i-2], f_array[i-1], func_alpa(p, s, (alpha_array[i-1] + alpha_array[i]) / 2),f_array[i]]

    if list4_f[1] < list4_f[2]:
        if direction == 1:
            return [list4_alpha[0], list4_alpha[2]]
        else:
            return [list4_alpha[2], list4_alpha[0]]
    else:
        if direction == 1:
            return [list4_alpha[1], list4_alpha[3]]
        else:
            return [list4_alpha[3], list4_alpha[1]]
```

```python
def golden_ratio(sv, p, s):
    global counter
    counter -= 2

    l = sv[1] - sv[0]
    l1 = sv[0]
    l2 = sv[0] + 0.382 * l
    l3 = sv[0] + 0.618 * l
    l4 = sv[1]

    f1 = func_alpa(p, s, l1)
    f2 = func_alpa(p, s, l2)
    f3 = func_alpa(p, s, l3)
    f4 = func_alpa(p, s, l4)

    while l4 - l1 > e_golden_ratio:
        if f1 > f4:
            f1 = f2
            l1 = l2
            f2 = f3
            l2 = l3

            l = l4 - l1
            l3 = l1 + 0.618 * l
            f3 = func_alpa(p, s, l3)
        else:
            f4 = f3
            l4 = l3
            f3 = f2
            l3 = l2

            l = l4 - l1
            l2 = l1 + 0.382 * l
            f2 = func_alpa(p, s, l2)
    return (l2 + l3) / 2
```

```python
def DSK_P(sv, p, s):
    global counter
    counter -= 3

    delta = (sv[1] - sv[0]) / 2
    l1 = sv[0]
    l2 = (sv[0] + sv[1]) / 2
    l3 = sv[1]
    f1 = func_alpa(p, s, l1)
    f2 = func_alpa(p, s, l2)
    f3 = func_alpa(p, s, l3)


    l = l2 + (delta * (f1 - f3) / (2 * (f1 - 2 * f2 + f3)))
    f = func_alpa(p, s, l)

    while True:
        if abs(l - l2) < e_golden_ratio and abs(f2 - f) < e_golden_ratio:
            return l
        if f2 > f:
            if l > l2:
                l1 = l2
                f1 = f2
            else:
                l3 = l2
                f3 = f2
            l2 = l
            f2 = f
        else:
            if l > l2:
                l3 = l
                f3 = f
            else:
                l1 = l
                f1 = f

        a1 = (f2 - f1) / (l2 - l1)
        a2 = 1 / (l3 - l2) * ((f3 - f1) / (l3 - l1) - (f2 - f1) / (l2 -l1))
        l = (l2 + l1) / 2 - a1 / 2 / a2
        f = func_alpa(p, s, l)
```

```python
def BFS():
    global counter
    counter = 0

    I = np.array([[1, 0], [0, 1]])
    A = I
    x = [np.array(x0)]
    k = 0

    while True:
        grad = d_center(x[k])
        S = -np.matmul(A, grad)

        sven = svenn(x[k], S)
        la_opt = DSK_P(sven, x[k], S)
        x_n = x[k] + la_opt * (S)
        x.append(x_n)

        grad_l = grad
        grad = d_center(x[k + 1])
        dg_T = grad - grad_l
        dg_T = np.array([dg_T])

        dx_T = x[k + 1] - x[k]
        dx_T = np.array([dx_T])

        dg = dg_T.transpose()
        dx = dx_T.transpose()

        counter -= 2
        if (norm_x(x[k] - x[k + 1]) / norm_x(x[k]) < e_bfsg)\
                and ((func([x[k][0], x[k][1]]) - func([x[k + 1][0], x[k + 1][1]])) < e_bfsg) \
                or norm_x(grad) < e_bfsg:
            break

        d1 = I - (np.matmul(dx, dg_T) / np.matmul(dx_T, dg))
        d2 = I - (np.matmul(dg, dx_T) / np.matmul(dx_T, dg))
        d3 = np.matmul(dx, dx_T) / np.matmul(dx_T, dg)
        A = np.matmul(np.matmul(d1, A), d2) + d3

        k += 1
    print("x_opt = ", x[-1])
    return x
```
